```r
library(ggplot2)

ggplot(data, aes(x = Wall.Area, y = Cooling.Load)) +
  geom_point() +
  labs(title = "Scatter Plot", x = "Wall Area", y = "Cooling Load")
```

```r
library(ggplot2)

transformed_data <- data %>%
  group_by(Relative.Compactness) %>%
  summarise(mean_cooling_load = mean(Cooling.Load))

ggplot(transformed_data, aes(x = Relative.Compactness, y = mean_cooling_load)) +
  geom_bar(stat = "identity") +
  labs(title = "Bar Chart", x = "Relative Compactness", y = "Mean Cooling Load")
```

```r
library(tidyr)
library(ggplot2)

line_data <- data%>%
  gather(key = "area_type", value = "area_value", Surface.Area)

ggplot(line_data, aes(x = area_value, y = Cooling.Load, color = area_type)) +
  geom_line() +
  labs(title = "Line Plot", x = "Area", y = "Cooling Load")
```
